# Since the data is highly unbalanced, we can separate the datasets to get better understanding

```python
legit = creditcard[creditcard.Class == 0]
fraud = creditcard[creditcard.Class == 1]
```

# As we can see that there is some pattern between legit and fradulent transactions 

# Since there are significantly more legitimate transactions than fraudulent ones, we need to under-sample the legitimate dataset to balance the classes. This will improve the chances of building a more accurate model by ensuring equal representation of fraudulent transactions in the dataset.

```python
legit_sample = legit.sample(n=492)
```

```python
new_creditcard = pd.concat([legit_sample, fraud], axis=0)
print(new_creditcard.Class.value_counts())
```

```python
new_creditcard.head()
```

```python
new_creditcard.tail()
```

```python
new_creditcard.groupby('Class').mean()
```

# Now the dataset is not disturbed statistically but value count of legit and fradulent transactions have changed and now the representation of both classes are equal.

# Now we split the input features from target variable and try to fit a Logistic Regression Model

```python
X = new_creditcard.drop('Class', axis=1)
Y = new_creditcard.Class
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=4)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

# Accuracy on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('The accuracy of the training data is', training_data_accuracy)
```

# Testing data accuracy

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('The accuracy of the test data is', test_data_accuracy)
```
